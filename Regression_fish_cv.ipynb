# Correlation Analysis with Heatmap

The heatmap is a powerful visualization tool that represents data values as colors, providing an overview of the data landscape in a compact format. In this context, we use a heatmap to visualize the correlation coefficients between pairs of variables in our dataset.

Correlation coefficients are numerical measures of the statistical relationship between two variables, ranging from -1 to 1. A coefficient of 1 indicates a perfect positive correlation (as one variable increases, so does the other), while -1 signifies a perfect negative correlation (as one variable increases, the other decreases). A coefficient of 0 implies no linear correlation between the variables.

## Interpreting the Heatmap

Each cell in the heatmap corresponds to the correlation coefficient between the variables denoted by its row and column. For instance, the cell at the intersection of the 'Species' row and the 'Length1' column represents the correlation between 'Species' and 'Length1'.

The diagonal cells, running from the top left to the bottom right, all have a value of 1, as these cells represent the correlation of each variable with itself.

The color scheme ranges from blue (negative correlation) to red (positive correlation), with white indicating no correlation (correlation coefficient close to 0).

Upon examining the heatmap, we observe that several variables, such as 'Length1', 'Length2', 'Length3', 'Height', 'Width', and 'Weight', exhibit strong positive correlations with each other (indicated in red). This suggests that these variables tend to increase or decrease in tandem. Conversely, the 'Species' variable appears to be negatively correlated with most of the other variables (indicated in blue), implying that as 'Species' increases, the other variables tend to decrease, and vice versa.

```python

#  'Weight' is the target variable
X = data.drop('Weight', axis=1)
y = data['Weight']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Define the models
models = {
    "linear_regression": LinearRegression(),
    "decision_tree": DecisionTreeRegressor(),
    "svm": SVR(),
    "knn": KNeighborsRegressor()
}

# Define the parameters for GridSearchCV
parameters = {
    "linear_regression": {},
    "decision_tree": {"max_depth": [None, 5, 10, 15, 20]},
    "svm": {"C": [0.1, 1, 10], "kernel": ["linear", "rbf"]},
    "knn": {"n_neighbors": [3, 5, 7, 9, 11]}
}

# Train the models and output the metrics
for model_name, model in models.items():
    print(f"Training {model_name}...")
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    print(f"Training set metrics for {model_name}:")
    print(f"R2: {r2_score(y_train, y_train_pred)}")
    print(f"MSE: {mean_squared_error(y_train, y_train_pred)}")
    print(f"RMSE: {np.sqrt(mean_squared_error(y_train, y_train_pred))}")
    print(f"MAE: {mean_absolute_error(y_train, y_train_pred)}")
    print(f"Test set metrics for {model_name}:")
    print(f"R2: {r2_score(y_test, y_test_pred)}")
    print(f"MSE: {mean_squared_error(y_test, y_test_pred)}")
    print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_test_pred))}")
    print(f"MAE: {mean_absolute_error(y_test, y_test_pred)}")
    grid_search = GridSearchCV(model, parameters[model_name], cv=5)
    grid_search.fit(X, y)
    print(f"Best parameters for {model_name}: {grid_search.best_params_}")

# Output the coefficients of the linear regression model
print(f"Coefficients of the linear regression model: {models['linear_regression'].coef_}")
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import pandas as pd

# Assuming that 'data' is  DataFrame
# Assuming that 'Weight' is the target variable
X = data.drop('Weight', axis=1)
y = data['Weight']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Define the models
models = {
    "linear_regression": LinearRegression(),
    "decision_tree": DecisionTreeRegressor(),
    "svm": SVR(),
    "knn": KNeighborsRegressor()
}

# Define the parameters for GridSearchCV
parameters = {
    "linear_regression": {},
    "decision_tree": {"max_depth": [None, 5, 10, 15, 20]},
    "svm": {"C": [0.1, 1, 10], "kernel": ["linear", "rbf"]},
    "knn": {"n_neighbors": [3, 5, 7, 9, 11]}
}

# Define the scalers
scalers = {
    "standard_scaler": StandardScaler(),
    "minmax_scaler": MinMaxScaler()
}

# Scale the data, train the models and output the metrics
for scaler_name, scaler in scalers.items():
    print(f"\nUsing {scaler_name}...")
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    for model_name, model in models.items():
        print(f"\nTraining {model_name}...")
        model.fit(X_train_scaled, y_train)
        y_train_pred = model.predict(X_train_scaled)
        y_test_pred = model.predict(X_test_scaled)
        print(f"Training set metrics for {model_name}:")
        print(f"R2: {r2_score(y_train, y_train_pred)}")
        print(f"MSE: {mean_squared_error(y_train, y_train_pred)}")
        print(f"RMSE: {np.sqrt(mean_squared_error(y_train, y_train_pred))}")
        print(f"MAE: {mean_absolute_error(y_train, y_train_pred)}")
        print(f"Test set metrics for {model_name}:")
        print(f"R2: {r2_score(y_test, y_test_pred)}")
        print(f"MSE: {mean_squared_error(y_test, y_test_pred)}")
        print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_test_pred))}")
        print(f"MAE: {mean_absolute_error(y_test, y_test_pred)}")
        grid_search = GridSearchCV(model, parameters[model_name], cv=5)
        grid_search.fit(X_train_scaled, y_train)
        print(f"Best parameters for {model_name}: {grid_search.best_params_}")

# Output the coefficients of the linear regression model
print(f"\nCoefficients of the linear regression model: {models['linear_regression'].coef_}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import pandas as pd


# Assuming that 'Weight' is the target variable
X = data.drop('Weight', axis=1)
y = data['Weight']

# Define the models with the best hyper-parameters found in task 2
models = {
    "linear_regression": LinearRegression(),
    "decision_tree": DecisionTreeRegressor(max_depth=20),
    "svm": SVR(C=10, kernel='linear'),
    "knn": KNeighborsRegressor(n_neighbors=3)
}

# Initialize the scaler
scaler = StandardScaler()

# Repeat the holdout 10 times
for i in range(10):
    print(f"\nHoldout {i+1}...")

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)

    # Scale the data
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Train the models and output the metrics
    for model_name, model in models.items():
        print(f"\nTraining {model_name}...")
        model.fit(X_train_scaled, y_train)
        y_train_pred = model.predict(X_train_scaled)
        y_test_pred = model.predict(X_test_scaled)
        print(f"Training set metrics for {model_name}:")
        print(f"R2: {r2_score(y_train, y_train_pred)}")
        print(f"MSE: {mean_squared_error(y_train, y_train_pred)}")
        print(f"RMSE: {np.sqrt(mean_squared_error(y_train, y_train_pred))}")
        print(f"MAE: {mean_absolute_error(y_train, y_train_pred)}")
        print(f"Test set metrics for {model_name}:")
        print(f"R2: {r2_score(y_test, y_test_pred)}")
        print(f"MSE: {mean_squared_error(y_test, y_test_pred)}")
        print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_test_pred))}")
        print(f"MAE: {mean_absolute_error(y_test, y_test_pred)}")
```

Create a dataframe to represent the results: rows correspond to the models, but columns to
the performance metrics (testset) of each holdout.

```python
# Initialize an empty dataframe to store the results
results_df = pd.DataFrame()

# Repeat the holdout 10 times
for i in range(10):
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)

    # Scale the data
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Train the models and store the metrics
    for model_name, model in models.items():
        model.fit(X_train_scaled, y_train)
        y_test_pred = model.predict(X_test_scaled)

        # Append the metrics to the results dataframe
        results_df = results_df.append({
            "Model": model_name,
            "Holdout": i+1,
            "R2": r2_score(y_test, y_test_pred),
            "MSE": mean_squared_error(y_test, y_test_pred),
            "RMSE": np.sqrt(mean_squared_error(y_test, y_test_pred)),
            "MAE": mean_absolute_error(y_test, y_test_pred)
        }, ignore_index=True)

# Pivot the dataframe so that rows correspond to the models and columns correspond to the performance metrics of each holdout
results_df = results_df.set_index(["Model", "Holdout"]).unstack()

# Print the results dataframe
print(results_df)
```
